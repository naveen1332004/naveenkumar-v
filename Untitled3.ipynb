```python
import numpy as np
import pickle as pkl
import tensorflow as tf
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.layers import GlobalMaxPool2D
from sklearn.neighbors import NearestNeighbors
import os
from numpy.linalg import norm
from IPython.display import Image

# Ensure the directory exists
os.makedirs('Fashion Recommendation', exist_ok=True)

# Collect image filenames
filenames = [os.path.join('images', file) for file in os.listdir('images')]
print(f"Number of images: {len(filenames)}")

# Load the ResNet50 model without the top layer for feature extraction
model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
model.trainable = False

# Add a GlobalMaxPool2D layer to reduce the output dimensions
model = tf.keras.models.Sequential([model, GlobalMaxPool2D()])
model.summary()

# Function to extract features from an image
def extract_features_from_images(image_path, model):
    img = image.load_img(image_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_expand_dim = np.expand_dims(img_array, axis=0)
    img_preprocess = preprocess_input(img_expand_dim)
    result = model.predict(img_preprocess).flatten()
    norm_result = result / norm(result)
    return norm_result

# Extract features for all images
image_features = [extract_features_from_images(file, model) for file in filenames[:5]]

# Save the extracted features and filenames using pickle
with open('Fashion Recommendation/Images_features.pkl', 'wb') as f:
    pkl.dump(image_features, f)

with open('filenames.pkl', 'wb') as f:
    pkl.dump(filenames, f)

# Load the extracted features and filenames from the saved files
image_features = pkl.load(open('Fashion Recommendation/Images_features.pkl', 'rb'))
filenames = pkl.load(open('filenames.pkl', 'rb'))

# Check the shape of image features
print(np.array(image_features).shape)

# Perform nearest neighbors search on the extracted features
input_image = extract_features_from_images('16871.jpg', model)  # Extract features of the input image

# Adjust n_neighbors to be the minimum of 6 or the number of available samples
n_neighbors = min(6, len(image_features))

neighbors = NearestNeighbors(n_neighbors=n_neighbors, algorithm='brute', metric='euclidean')
neighbors.fit(image_features)  # Fit the model with image features
distance, indices = neighbors.kneighbors([input_image])  # Find the nearest neighbors

# Display the indices of the nearest neighbors
print(f"Indices of nearest neighbors: {indices[0]}")

# Display the nearest images
print("Nearest images:")
for idx in indices[0]:
    print(f"Image: {filenames[idx]}")
    display(Image(filename=filenames[idx]))
    
```
